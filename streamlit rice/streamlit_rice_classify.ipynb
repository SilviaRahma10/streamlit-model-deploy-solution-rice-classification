```python

%%writefile app.py
import streamlit as st
import pandas as pd
from PIL import Image
import streamlit as st


import streamlit as st
import pandas as pd
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
from tensorflow.keras import preprocessing
from tensorflow.keras.models import load_model
from tensorflow.keras.activations import softmax
import os
import h5py


def main_page():
    st.sidebar.markdown("###### Ruce Classification")
    with st.container():
        st.markdown('---')
        st.markdown\
             ("<h1 style='text-align: center; color: #728FCE;'>CLASSIFICATIN RICE</h1>", unsafe_allow_html=True)
        st.markdown("<h3 style='text-align: center; color: #728FCE;'>(by Silvia Rahma)</h1>", unsafe_allow_html=True)
        st.markdown("<h1 style='text-align: center; color: #728FCE;'>ğŸ–¥ï¸ğŸŒ¾ğŸ§‘</h1>", unsafe_allow_html=True)
        st.markdown('---')

        st.markdown("<h3 style='color: #728FCE;'>Rice, which is among the most widely produced grain products worldwide, has many genetic varieties. These varieties are separated from each other due to some of their features. These are usually features such as texture, shape, and color. With these features that distinguish rice varieties, it is possible to classify and evaluate the quality of seeds.</h3>", unsafe_allow_html=True)


        st.markdown("<h3 style='color: #728FCE;'>In this study there are :</h3>", unsafe_allow_html=True)
        st.markdown('- Arborio'
            '\n- Basmati'
            '\n- Ipsala'
            '\n- Jasmine'
            '\n- Karacadag')

# alzheimer
def rice():
    st.sidebar.markdown('- Rice Classification')

    st.markdown('---')
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>Rice Classification</h1>", unsafe_allow_html=True)
    st.markdown("<h1 style='text-align: center; color: #728FCE;'>ğŸ¦ </h1>", unsafe_allow_html=True)
    st.markdown('---')

    # Deteksi penyakit
    st.markdown("<h2 style='color: #728FCE;'>Upload a picture of rice</h1>", unsafe_allow_html=True)

        #fungsi untuk prediksi kelas
    def predict_class_rice(image):
        classifier_model = tf.keras.models.load_model(r'/content/drive/MyDrive/portofolio/streamlit rice/Hasil/model_inception_rice.h5')
        shape = ((416, 416, 3))
        model = tf.keras.Sequential([hub.KerasLayer(classifier_model, input_shape=shape)])

        test_image = image.resize((416, 416))
        test_image = preprocessing.image.img_to_array(test_image)
        #test_image = test_image/225.0
        test_image = np.expand_dims(test_image, axis = 0)
        class_names = ['Arborio', 'Basmati', 'Ipsala', 'Jasmine', 'Karacadag']
        prediction = model.predict(test_image)
        score = tf.nn.softmax(prediction[0])
        score = score.numpy()
        image_class =  class_names[np.argmax(score)]
        return image_class

 # fungsi untuk import atau upload gambar
    file_uploaded = st.file_uploader("Pilih File", type = ['jpg', 'png', 'jpeg'])
    if file_uploaded is not None:
        image = Image.open(file_uploaded)
        figure = plt.figure(figsize=(4, 3))
        plt.imshow(image)
        plt.axis('off')
        predict_class = predict_class_rice(image)
        hasil = "Rice Type : {}".format(predict_class)
        st.write(hasil)
        st.pyplot(figure)


page_names_to_funcs = {
    "Headline": main_page,
    "Rice Classification": rice,
}

selected_page = st.sidebar.selectbox("Select a page", page_names_to_funcs.keys())
page_names_to_funcs[selected_page]()
```
